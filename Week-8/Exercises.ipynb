# Visualizing Company Sale data

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

### Step 2. Import the dataset from this [address](https://github.com/thieu1995/csv-files/blob/main/data/matplotlib/company_sales_data.csv).

```python
df = pd.read_csv("https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv")
df
```

### Exercise 1: Read Total profit of all months and show it using a line plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/01.png?raw=1)

```python
fig, ax = plt.subplots(figsize= (11,6))
ax.plot(df.month_number, df.total_profit)
ax.set( xlabel="Month number", ylabel= "Total Profit", title= "Company profit per month", 
       xticks=range(1,13),  yticks= np.linspace(100000,500000,5));
```

### Exercise 2: Get total profit of all months and show line plot with the following Style properties

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/02.png?raw=1)

```python
fig, ax = plt.subplots(figsize= (11,6))
ax.plot(df.month_number, df.total_profit, 
        linestyle="--", color="red", linewidth= 6,
        marker="o",markersize= 12, markerfacecolor="black", markeredgecolor="red", markeredgewidth=2.5,
        label="Profit data of last year")

ax.set(xlabel="Month number", ylabel= "Total Profit", 
       title= "Company profit per month", 
       xticks=range(1,13),  yticks= np.linspace(100000,500000,5));
ax.legend();
```

### Exercise 3: Read all product sales data and show it  using a multiline plot

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/03.png?raw=1)

```python
fig, ax = plt.subplots(figsize= (11,6))
ax.plot(df.month_number, df.facecream, 
        color="blue", linewidth= 6,
        marker="o",markersize= 12, 
        label="Face cream Sales Data");

ax.plot(df.month_number, df.facewash, 
        color="orange", linewidth= 6,
        marker="o",markersize= 12, 
        label="Face Wash Sales Data");

ax.plot(df.month_number, df.toothpaste, 
        color="green", linewidth= 6,
        marker="o",markersize= 12,
        label="TooththPaste Sales Data");

ax.plot(df.month_number, df.bathingsoap, 
        color="red", linewidth= 6,
        marker="o",markersize= 12,
        label="Bathingsoap Sales Data");

ax.plot(df.month_number, df.shampoo, 
        color="violet", linewidth= 6,
        marker="o",markersize= 12,
        label="Shampoo Sales Data");

ax.plot(df.month_number, df.moisturizer, 
        color="brown", linewidth= 6,
        marker="o",markersize= 12, 
        label="Moisturizer Sales Data");

ax.set(xlabel="Month number", ylabel= "Sales unnits in number", 
       title= "Company profit per month", 
       xticks=range(1,13),  yticks= [1000,2000,4000,6000,8000,10000,12000,15000,18000]);
ax.legend();
```

### Exercise 4: Read toothpaste sales data of each month and show it using a scatter plot
![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/04.png?raw=1)

```python

fig, ax = plt.subplots(figsize= (11,6))
ax.scatter(df.month_number, df.toothpaste, linewidths= 5, label= "Tooth paste Sales data");
ax.set(xlabel="Month number", ylabel= "Number of units Sold", 
       title= "Company profit per month", 
       xticks=range(1,13));
ax.grid(True, linestyle="--", linewidth=2)
ax.legend();
```

### Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

![img](https://github.com/thieu1995/pandas/blob/main/11_matplotlib/company-sale/img/05.png?raw=1)

```python
fig, ax = plt.subplots(figsize=(11,6))
ax.set(xlabel="Month number", ylabel= "Sales unit in number", 
       title= "Facewash and facecream sales data", 
       xticks=range(1,13));
ax.grid(True, linestyle="--", linewidth=2);
width = 0.35
ax.bar(df.month_number - width/2, df.facecream, width=width, label='Facecream Sales data')
ax.bar(df.month_number + width/2, df.facewash, width=width, label='Facewash Sales data')

ax.legend();
```
